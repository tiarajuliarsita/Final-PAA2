```python
import time
```

```python

def bubble_sort(arr):
    n = len(arr)
    start_time = time.time()

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

        if i < 5 or i >= n-5:
            print(f"Iteration {i+1}: {arr}")

    end_time = time.time()
    execution_time = end_time - start_time

    return arr, execution_time
```

```python
def insertion_sort(arr):
    n = len(arr)
    start_time = time.time()

    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

        if i < 5 or i >= n-5:
            print(f"Iteration {i+1}: {arr}")

    end_time = time.time()
    execution_time = end_time - start_time

    return arr, execution_time
```

###  **B. Menghitung shortest path menggunakan algoritma TSP dan Djikstra**

```python
import time
import sys
```

```python


# Fungsi untuk menghitung shortest path menggunakan algoritma TSP
def tsp(graph, start):
    # Menentukan semua permutasi jalur yang mungkin
    cities = list(graph.keys())
    cities.remove(start)
    permutations = get_permutations(cities)
    shortest_path = sys.maxsize

    for path in permutations:
        current_path = [start] + path + [start]
        total_distance = 0

        # Menghitung total jarak pada jalur saat ini
        for i in range(len(current_path) - 1):
            current_city = current_path[i]
            next_city = current_path[i + 1]
            if next_city in graph[current_city]:
                total_distance += graph[current_city][next_city]
            else:
                total_distance = sys.maxsize
                break

        # Memperbarui shortest path jika ditemukan jalur yang lebih pendek
        if total_distance < shortest_path:
            shortest_path = total_distance
            shortest_path_order = current_path

        # Menampilkan hasil setiap iterasi
        print("Iterasi:", permutations.index(path) + 1)
        print("Jalur:", current_path)
        print("Jarak:", total_distance)
        print()

    # Menampilkan hasil akhir shortest path
    print("Hasil Akhir:")
    print("Shortest Path:", shortest_path_order)
    print("Jarak Terpendek:", shortest_path)

# Fungsi untuk mendapatkan semua permutasi jalur yang mungkin
def get_permutations(cities):
    if len(cities) == 1:
        return [cities]

    permutations = []
    for i in range(len(cities)):
        current_city = cities[i]
        remaining_cities = cities[:i] + cities[i+1:]
        for perm in get_permutations(remaining_cities):
            permutations.append([current_city] + perm)

    return permutations
```

```python

# Fungsi untuk menghitung shortest path menggunakan algoritma Dijkstra
def dijkstra(graph, start):
    # Inisialisasi variabel yang diperlukan
    distances = {}
    visited = {}
    previous = {}
    for city in graph:
        distances[city] = sys.maxsize
        visited[city] = False
        previous[city] = None
    distances[start] = 0

    while True:
        # Memilih kota dengan jarak terpendek yang belum dikunjungi
        current_city = None
        min_distance = sys.maxsize
        for city in graph:
            if not visited[city] and distances[city] < min_distance:
                min_distance = distances[city]
                current_city = city

        if current_city is None:
            break

        # Memperbarui jarak ke tetangga-tetangga yang belum dikunjungi
        for neighbor in graph[current_city]:
            distance = distances[current_city] + graph[current_city][neighbor]
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_city

        visited[current_city] = True

        # Menampilkan hasil setiap iterasi
        print("Iterasi:", len(visited) - list(visited.values()).count(False))
        print("Distances:", distances)
        print("Previous:", previous)
        print()

    # Mendapatkan shortest path dari start ke setiap kota lainnya
    shortest_path = {}
    for city in graph:
        path = get_shortest_path(previous, start, city)
        shortest_path[city] = {"Path": path, "Distance": distances[city]}

    # Menampilkan hasil akhir shortest path
    print("Hasil Akhir:")
    for city in shortest_path:
        print("Kota Tujuan:", city)
        print("Shortest Path:", shortest_path[city]["Path"])
        print("Jarak Terpendek:", shortest_path[city]["Distance"])
        print()

# Fungsi untuk mendapatkan shortest path dari start ke destination
def get_shortest_path(previous, start, destination):
    path = [destination]
    current_city = destination
    while current_city != start:
        current_city = previous[current_city]
        path.append(current_city)
    path.reverse()
    return path
```

```python

# Fungsi utama
def main():
    # Membangun graph
    graph = {
        "A": {"B": 12, "G": 12, "C": 10},
        "B": {"A": 12, "C": 8, "D": 12},
        "C": {"B": 8, "D": 11, "E": 3, "G": 9, "A": 10},
        "D": {"C": 11, "E": 11, "B": 12, "F": 10},
        "E": {"D": 11, "F": 6, "C": 3, "G": 7},
        "F": {"D": 10, "E": 6, "G": 9},
        "G": {"A": 12, "E": 7, "F": 9, "C": 9}
    }

    while True:
        print("\n\nPilih algoritma:")
        print("1. TSP (Traveling Salesman Problem)")
        print("2. Dijkstra")
        print("3. Keluar")
        choice = input("Masukkan pilihan (1/2/3): ")

        if choice == '1':
            start_time = time.time()
            tsp(graph, "A")
            end_time = time.time()
            computation_time = end_time - start_time
            print("Waktu komputasi: %.6f detik" % computation_time)
        elif choice == '2':
            start_time = time.time()
            dijkstra(graph, "A")
            end_time = time.time()
            computation_time = end_time - start_time
            print("Waktu komputasi: %.6f detik" % computation_time)
        elif choice == '3':
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
